## 测试 openai 接口调用deepseek

```python
api_key = ''
```

```python
import os
from utils.llm_client.openai import OpenAIClient

# DeepSeek API Key 和 Base URL
base_url = "https://api.deepseek.com/v1"  # DeepSeek 的 API 地址

# 初始化 DeepSeek 客户端（使用 OpenAIClient，替换 base_url）
client = OpenAIClient(
    model="deepseek-chat",  # DeepSeek 支持的模型名
    temperature=0.7,
    base_url=base_url,
    api_key=api_key,
)

# 构造消息
messages = [
    {"role": "system", "content": "你是一个有用的助手。"},
    {"role": "user", "content": "你好，帮我写一个快速排序的Python代码。"}
]

# 调用 chat completion
response = client._chat_completion_api(messages, temperature=0.7, n=1)
print("DeepSeek 返回内容：")
print(response[0].message.content)
```
