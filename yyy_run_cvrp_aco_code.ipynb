```python
# 设置参数

import os

root_dir = os.getcwd()
eval_file_path = root_dir + '/problems' + '/cvrp_aco' + '/eval.py'
problem_size = 50
output_file = root_dir + '/problems' + '/cvrp_aco' + '/gpt.py'

outputs_dir = 'F:\\reevo\\outputs\\cvrp_aco-aco\\2025-10-25_22-25-51\\'
```

```python
# read log data
import json

with open(outputs_dir + 'analysis_log.json', 'r') as f:
    log_data = json.load(f)
```

```python
def get_ind_by_code_path(code_path):
    for ind in log_data['population']:
        if ind['code_path'] == code_path:
            return ind
    return None
```

### 根据 iter number& response number 运行个体

example (2，26)


```python
def read_code_by_id(iter_num, resp_num):
    with open(outputs_dir + f'problem_iter{iter_num}_response{resp_num}.txt') as f:
        code = ''.join(f.readlines()[1:-1])
    # print(code)
    return code
```

### 运行字符串形式的代码

```python
import sys
import subprocess
import os
import time

def run_str_code(code):
    
    # 写入文件
    with open(output_file, 'w') as file:
        file.write(code + '\n')
    
    # 强制刷新并等待一小段时间确保文件写入完成
    file_handle = open(output_file, 'r')
    file_handle.flush()
    file_handle.close()
    time.sleep(0.1)  # 短暂等待确保文件写入
    
    # 运行子进程
    process = subprocess.Popen(
        [sys.executable, '-u', eval_file_path, f'{problem_size}', root_dir, "train"],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True
    )
    
    # 读取并打印输出
    for line in iter(process.stdout.readline, ''):
        print(line.rstrip())
    
    process.wait()
    return process.returncode
```
